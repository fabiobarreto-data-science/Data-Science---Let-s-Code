## cut

```python
arr = np.array([1, 7, 5, 4, 6, 3])
```

```python
pd.cut(x=arr, bins=3) # Separar os dados em intervalos discretos. Separa eles em dados categóricos ou hierarquicos. 'bins=3' classifca em 3 categorias ou hierarquias.
```

```python
labels = ['bom', 'medio', 'mal']
pd.cut(x=arr, bins=3, labels=labels)
```

```python
pd.cut(x=arr, bins=3, labels=labels).tolist() # Transformando em lista
```

```python
pd.cut(x=arr, bins=3, labels=False) # Ele substitui os labels por números
```

## pivot_table

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7]})
df
```

```python
def recebe_funcao_aplica(func, x):
    return func(x)

def soma(x):
    return sum(x)
```

```python
recebe_funcao_aplica(soma, [2,3])
```

```python
pd.pivot_table(data=df, index='A', columns='C', values='D', aggfunc=soma)
```

```python
pd.pivot_table(data=df, index='A', columns='C', values='D', aggfunc=np.median)
```
