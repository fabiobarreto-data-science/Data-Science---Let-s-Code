Enunciado
Crie um array x com 200 pontos, começando de -10 até 10. Depois calcule as seguintes funções sobre esse array:

y1 = sen(x) / x
y2 = 1 / (1-e^(-x))

Por fim, plot essas duas funções, em função do x, ambas em uma mesma figura, porém, em dois axes distintos (um abaixo do outro).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 200)
```

```python
e = 2.71
y1 = np.sin(x) / x 
y2 = 1 / (1-e**(-x))
```

```python
fig, axs = plt.subplots(2);
fig.suptitle("Dois gráficos");

axs[0].set_title("Gráfico 01")
axs[0].plot(y1, color="green")

axs[1].set_title("Gráfico 02")
axs[1].plot(y2, color="yellow")
```
