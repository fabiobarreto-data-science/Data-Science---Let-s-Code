## subplots

```python
import numpy as np

# Dados de exemplo
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y);
ax.set_title("Seno x²");
```

E se eu quiser adicionar um gráfico em baixo do meu primeiro gráfico?

```python
fig, axs = plt.subplots(2)
fig.suptitle("Dois gráficos")

axs[0].set_title("Grafico 1")
axs[0].plot(x, y, color="red")

axs[1].set_title("Grafico 2")
axs[1].plot(x, -y, color="orange");
```

Mas... E se eu não quiser essa sobreposição?

```python
fig, axs = plt.subplots(nrows=2, ncols=1)
fig.suptitle('Dois gráficos')

axs[0].set_title('Grafo 1')
axs[0].plot(x, y, color='red')

axs[1].set_title('Grafo 2')
axs[1].plot(x, -y, color='orange')

plt.tight_layout();
```

Como eu faço para ter mais de uma coluna?

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

ax1.plot(x, y, color='red')
ax2.plot(x, -y, color='orange') 

plt.tight_layout()
```

E se a gente quiser gerar 4 gráficos: onde temos 2 linhas e 2 colunas:

```python
fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].set_title('grafico [0, 0]')
axs[0, 0].plot(x, y)

axs[0, 1].set_title('grafico [0, 1]')
axs[0, 1].plot(x, -y, color='red')

axs[1, 0].set_title('grafico [1, 0]')
axs[1, 0].plot(x, y**2, color='orange')

axs[1, 1].set_title('grafico [1, 1]')
axs[1, 1].plot(x, y**3, color='green')

plt.tight_layout()
```

Para guiá-los na escolha das cores:

* https://matplotlib.org/stable/tutorials/colors/colors.html

```python
fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].set_title('grafico [0, 0]')
axs[0, 0].plot(x, y)

axs[0, 1].set_title('grafico [0, 1]')
axs[0, 1].plot(x, -y, color='red')

axs[1, 0].set_title('grafico [1, 0]')
axs[1, 0].plot(x, y**2, color='orange')

axs[1, 1].set_title('grafico [1, 1]')
axs[1, 1].plot(x, y**3, color='green')

plt.tight_layout()

# axs.flat: transforma o meu axs em um array de uma dimensão shape=(n, 1)
for ax in axs.flat:
    ax.set_title('Teste');
```

Iterar por todos os items:

```python
fig, axs = plt.subplots(nrows=2, ncols=3)

for index_0 in range(axs.shape[0]):
    for index_1 in range(axs.shape[1]):
        axs[index_0, index_1].set_title('grafico ['  +str(index_0) + ', ' + str(index_1) + ']')
        axs[index_0, index_1].plot(x, y)
        
plt.tight_layout();
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3)

for indice, ax in np.ndenumerate(axs):
    ax.set_title(f'Grafico {indice}')
    ax.plot(x, y)
    
plt.tight_layout()
```

E se quisesse um gráfico na primeira linha e dois na segunda:

```python
gridsize = (2, 2)

ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2)
ax2 = plt.subplot2grid(gridsize, (1, 0))
ax3 = plt.subplot2grid(gridsize, (1, 1))

ax1.plot(x, y, color='red')
ax2.plot(x, y, color='orange')
ax3.plot(x, y)

plt.tight_layout()
```
