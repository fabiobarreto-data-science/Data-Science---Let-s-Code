```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

(a) Crie um gráfico de pizza que apresente o percentual dos gêneros (Genres) dos aplicativos. Como existem muitos gêneros, considere apenas os 10 primeiros.

```python
app = pd.read_csv("C:\\Users\\Fabio\\Desktop\\Data Science\\Módulo 3\\googleplaystore.csv")
app_10 = app.head(10)
app_10
```

```python
plt.rcParams['figure.figsize']=15,5 # Parâmetro usado para esticar e/ou aumentar o gráfico. Esta única referência serve para todos os gráficos
```

```python
app_10["Genres"].value_counts()
plt.pie(app_10["Genres"].value_counts(), labels=app_10["Genres"].value_counts().index, autopct="%1.2f%%", shadow=True, startangle=90);
```

(b) Crie um gráfico de barras que contenha a quantidade de aplicativos que foram desenvolvidos para cada versão específica (ou conjunto de versões) do Android, especificadas na coluna Android Ver.

```python
app_filtro = app.groupby(by=["Android Ver"]).count()
app_filtro = app_filtro[app_filtro["App"]> 10]

grafico = plt.barh(app_filtro.index, app_filtro["App"], color="blue");
plt.title("Quantidade de aplicativos desenvolvida por versão do Android")
plt.xticks();
```
