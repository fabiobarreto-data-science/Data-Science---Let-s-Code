Carregue o arquivo houses_to_rent.csv. Em seguida:

(a) Crie um gráfico de dispersão para visualizar uma relação entre a área e o preço total do aluguel do imóvel.

(b) Visualize os outliers presentes na coluna da área utilizando o boxplot.

(c) Faça o mesmo que foi solicitado no item anterior, porém, para a coluna do preço total do aluguel.

(d) Elimine os outliers de ambas as colunas utilizando o método quantile. (e) Crie novamente um gráfico de dispersão dos dados da área e preço total do aluguel do imóvel; dessa vez, após a remoção dos outliers de ambas as colunas.

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
casas_para_vender = pd.read_csv("C:\\Users\\Fabio\\Desktop\\Data Science\\Módulo 3\\houses_to_rent.csv")
casas_para_vender.head(5)
```

```python
plt.rcParams['figure.figsize']=15,5 # Parâmetro usado para esticar e/ou aumentar o gráfico. Esta única referência serve para todos os gráficos
```

(a) Crie um gráfico de dispersão para visualizar uma relação entre a área e o preço total do aluguel do imóvel.

```python
plt.scatter(x=casas_para_vender["total (R$)"], y=casas_para_vender["area"], marker="*")
plt.title("Relação entre a área e o preço total do aluguel do imóvel")
plt.xlabel("total (R$)")
plt.ylabel("Área")
plt.show();
```

(b) Visualize os outliers presentes na coluna da área utilizando o boxplot.

```python
plt.boxplot(x=casas_para_vender['area'])
plt.title('Outliers - Área')
plt.ylabel('Área')
plt.show();
```

(c) Faça o mesmo que foi solicitado no item anterior, porém, para a coluna do preço total do aluguel.

```python
plt.boxplot(x=casas_para_vender["total (R$)"])
plt.title("Outliers - total")
plt.ylabel("Total (R&)")
plt.show();
```

(d) Elimine os outliers de ambas as colunas utilizando o método quantile.

```python
p = casas_para_vender["total (R$)"].quantile(0.99)
q = casas_para_vender["area"].quantile(0.99)
sem_outliers = casas_para_vender[(casas_para_vender["total (R$)"] < p) & (casas_para_vender["area"] < q)]
```

(e) Crie novamente um gráfico de dispersão dos dados da área e preço total do aluguel do imóvel; dessa vez, após a remoção dos outliers de ambas as colunas.

```python
plt.scatter(x=sem_outliers["total (R$)"], y=sem_outliers["area"])
plt.title("Área de casas disponíveis para alugar por valor total do aluguel")
plt.xlabel("Aluguel (R$)")
plt.ylabel("Área")
plt.show();
```
